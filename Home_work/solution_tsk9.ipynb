## Генерация сигналов

```python
ts_num = 10

# индекс и отсчет времени в секундах
time = np.linspace(0, 10, 100)

rand = np.random.random(size=100)

sin_ts = []
cos_ts = []

for i in range(ts_num):
    sin_ts.append(1.5 * np.sin(time +  rand[i] + rand[i**2]) + np.random.random())
    cos_ts.append(0.7 * np.cos(time +  rand[i] + rand[i**2]) + np.random.random())
```

```python
for i in range(ts_num):
    if (i == 0):
        plt.plot(time, sin_ts[i], '-b', label="Синусоидальный")
        plt.plot(time, cos_ts[i], '-r', label="Косинусоидальный")
    else:
        plt.plot(time, sin_ts[i], '-b')
        plt.plot(time, cos_ts[i], '-r')
  
            
plt.title('Сигналы')
plt.legend()
plt.xlabel('Время, с')
plt.ylabel('Амплитуда')
plt.grid()
```

## Shapelet Func

```python
from scipy.spatial.distance import euclidean

def distances_func(S, Wi_l):
    distances = [euclidean(S, candidate) for candidate in Wi_l]
    return distances

def find_candidates(ts, min_l = 3, max_l = 100):
    cands = []
    for l in range(min_l, max_l + 1):
        for i in range(len(ts) - l + 1):
            sub_sequence = ts[i:i + l]
            cands.append(sub_sequence)
    return cands

def cand_func(dist):
    return np.mean(np.sqrt(dist))


def insert_shapelet(rShapelets, shapelet):
    result = []
    i = 0
    for i in range(len(rShapelets)):
        if rShapelets[i][1] <= shapelet[1]:
            result += [rShapelets[i]]
        if rShapelets[i][1] > shapelet[1]:
            if i == 0 or rShapelets[i - 1][1] < shapelet[1]:
                result += [shapelet]
            else:
                result += [rShapelets[i - 1]]
    return result


def merge(k, kShapelets, shapelets):
    if len(kShapelets) == 0:
        return shapelets[0:k]
    result = kShapelets
    for shapelet in shapelets:
        if shapelet[1] < kShapelets[len(kShapelets) - 1][1]:
            kShapelets = insert_shapelet(result, shapelet)
    return result

def filter_shapelets(x, min = 0.05):
    result = []
    for xi in x:
        if xi[1] <= min:
            result += [xi]
    return result

def shapelet_selection(tseries_list, k = 50, min_len = 3, max_len = 100):
    kShapelets = []
    for tseries in tseries_list:
        shapelets = []
        for l in range(min_len, max_len + 1):
            candidates = find_candidates(tseries, min_len, max_len)
            for candidate in candidates:
                distances = distances_func(candidate, candidates)
                quality = cand_func(distances)
                shapelets.append((candidate, quality))
        shapelets = filter_shapelets(shapelets, quality)
        shapelets.sort(key=lambda x: x[1], reverse=True)
        kShapelets = merge(k, kShapelets, shapelets)
    return kShapelets
```
