## Генерация временных рядов

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
ts_num = 7

# индекс и отсчет времени в секундах
time = np.linspace(0, 10, 100)

rand = np.random.random(size=100)

sin_ts = []
cos_ts = []
normal_ts = []

for i in range(ts_num):
    sin_ts.append(1.5 * np.sin(time +  rand[i] + rand[i**2]) + np.random.random())
    cos_ts.append(0.7 * np.cos(time +  rand[i] + rand[i**2]) + np.random.random())

    mean = random.uniform(0.3, 1)
    std = random.uniform(0.5, 1.5)
    series = np.random.normal(mean, std, size=time.size)
    normal_ts.append(series)
```

```python
for i in range(ts_num):
    if (i == 0):
        plt.plot(time, sin_ts[i], '-b', label="Синусоидальный")
        plt.plot(time, cos_ts[i], '-r', label="Косинусоидальный")
        plt.plot(time, normal_ts[i], '-g', label="Нормальный")
    else:
        plt.plot(time, sin_ts[i], '-b')
        plt.plot(time, cos_ts[i], '-r')
        plt.plot(time, normal_ts[i], '-g')
  
            
plt.title('Сигналы')
plt.legend()
plt.xlabel('Время, с')
plt.ylabel('Амплитуда')
plt.grid()
```

## Подсчет расстояний

```python
from scipy.spatial.distance import euclidean

for i in range(1, 4):
    x = np.abs(sin_ts[i])
    y = np.abs(sin_ts[ts_num - i])
    distance = euclidean(x, y)
    print(f"euclidean sin {i} и {ts_num - i}:", distance)

    x = np.abs(cos_ts[i])
    y = np.abs(cos_ts[ts_num - i])
    distance = euclidean(x, y)
    print(f"euclidean для cos {i} и {7 - i}:", distance)

    x = np.abs(normal_ts[i])
    y = np.abs(normal_ts[ts_num - i])
    distance = euclidean(x, y)
    print(f"euclidean для normal {i} и {7 - i}:", distance)


x = np.abs(sin_ts[0])
y = np.abs(cos_ts[0])
distance = euclidean(x, y)
print(f"euclidean для sin_1 и cos_1:", distance)

x = np.abs(sin_ts[0])
y = np.abs(normal_ts[0])
distance = euclidean(x, y)
print(f"euclidean для sin_1 и normal_1:", distance)

x = np.abs(cos_ts[0])
y = np.abs(normal_ts[0])
distance = euclidean(x, y)
print(f"euclidean для cos_1 и normal_1:", distance)
```

```python
from dtaidistance.dtw import distance as dtw


for i in range(1, 4):
    x = np.abs(sin_ts[i])
    y = np.abs(sin_ts[ts_num - i])
    distance = dtw(x, y)
    print(f"euclidean sin {i} и {ts_num - i}:", distance)

    x = np.abs(cos_ts[i])
    y = np.abs(cos_ts[ts_num - i])
    distance = dtw(x, y)
    print(f"euclidean для cos {i} и {7 - i}:", distance)

    x = np.abs(normal_ts[i])
    y = np.abs(normal_ts[ts_num - i])
    distance = dtw(x, y)
    print(f"euclidean для normal {i} и {7 - i}:", distance)


x = np.abs(sin_ts[0])
y = np.abs(cos_ts[0])
distance = dtw(x, y)
print(f"dtw для sin_1 и cos_1:", distance)

x = np.abs(sin_ts[0])
y = np.abs(normal_ts[0])
distance = dtw(x, y)
print(f"dtw для sin_1 и normal_1:", distance)

x = np.abs(cos_ts[0])
y = np.abs(normal_ts[0])
distance = dtw(x, y)
print(f"dtw для cos_1 и normal_1:", distance)
```

dtw расстояние меньше, чем евклидово, для всех ts: как внутри класса, так и для ts в другими классами. 

## Наборы данных для обучения

```python
class_nums = []
class_data = []

for idx in range(ts_num):
    class_data.append(sin_ts[idx])
    class_nums.append(1)
    
for idx in range(ts_num):
    class_data.append(cos_ts[idx])
    class_nums.append(2)

for idx in range(ts_num):
    class_data.append(normal_ts[idx])
    class_nums.append(3)
```

```python
arr = np.arange(len(class_nums))
np.random.shuffle(arr)

X_train = []
X_test = []
Y_train = []
Y_test = []

for i in arr[:-6]:
    X_train.append(class_data[i])
    Y_train.append(class_nums[i])
for i in arr[len(X_train):]:
    X_test.append(class_data[i])
    Y_test.append(class_nums[i])

Y_train, Y_test 
```

## Обучение

```python
# dtw метрика
from pyts.classification import KNeighborsClassifier

clf = KNeighborsClassifier(metric='dtw')
clf.fit(X_train, Y_train)

print(f"Accuracy {clf.score(X_test, Y_test)}")
for i in range(len(Y_test)):
    print(f"{i + 1} объект:")
    print(f"Вектор вероятности принадлежности к классам {clf.predict_proba(X_test[i].reshape(1, -1))}")
    print(f"Истинный класс для предсказаний {Y_test[i]}")
```

```python
# стандартная метрика
clf = KNeighborsClassifier()
clf.fit(X_train, Y_train)

print(f"Accuracy {clf.score(X_test, Y_test)}")
for i in range(len(Y_test)):
    print(f"{i + 1} объект:")
    print(f"Вектор вероятности принадлежности к классам {clf.predict_proba(X_test[i].reshape(1, -1))}")
    print(f"Истинный класс для предсказаний {Y_test[i]}")
     
```

```python
# лес
from pyts.classification import TimeSeriesForest
 
clf = TimeSeriesForest(random_state=43)
clf.fit(X_train, Y_train)


print(f"Accuracy {clf.score(X_test, Y_test)}")
for i in range(len(Y_test)):
    print(f"{i + 1} объект:")
    print(f"Вектор вероятности принадлежности к классам {clf.predict_proba(X_test[i].reshape(1, -1))}")
    print(f"Истинный класс для предсказаний {Y_test[i]}")
     
```
